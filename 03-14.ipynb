```python
intersect = np.empty(len(df), dtype=np.csingle)
```

```python
for i in range(len(df)):
    if df.loc[i]["m_e"] == -1:
        val = complex(df.loc[i]["b_e"], df.loc[i]["b_w"])
        intersect[i] = val
    elif df.loc[i]["m_w"] == -1:
        val = complex(df.loc[i]["b_w"], df.loc[i]["b_e"])
        intersect[i] = val
    else:
        x = linalg.div_real(int(df.loc[i]["b_w"]-df.loc[i]["b_e"]), int(df.loc[i]["m_e"]-df.loc[i]["m_w"]))
        y = linalg.sim_s(df.loc[i]["m_e"] * x + df.loc[i]["b_e"])
        intersect[i] = complex(x,y)
```

```python
df["intersect"] = intersect
```

```python
import matplotlib.pyplot as plt

for i in range(1900, 1920):
    plt.figure(figsize=(7,7))
    points = [complex(df.loc[i]["eigval1"]), complex(df.loc[i]["eigval2"])]
    x = np.array([elem.real for elem in points], dtype=int)
    y = np.array([elem.imag for elem in points], dtype=int)
    plt.scatter(x,y, marker="s", c='b', s=200)
    avg_x = linalg.div_real(int(x.sum()), 2)
    avg_y = linalg.div_real(int(y.sum()), 2)
    plt.scatter(avg_x, avg_y, marker="v", c='darkorange', s=200)
    plt.scatter(complex(df.loc[i]["intersect"]).real, complex(df.loc[i]["intersect"]).imag, marker="^", c='g', s=200)
    plt.axis([0, 6, 0, 6])
    plt.show()
```
